```python
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head()
```

```python
df_can.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'}, inplace=True)
df_can.columns
```

```python
df_can['Total'] = df_can.sum(axis=1)
df_can.head(2)
```

```python
df_can.set_index('Country', inplace=True)
df_can.index.name=None #To remove name of index
df_can.head(2)
```

```python
df_can.loc['Japan', [1980, 1981, 1982, 1983, 1984, 1984]]
```

```python
df_can.columns = list(map(str, df_can.columns))
df_can.head(2)
```

```python
years = list(str(i) for i in range(1980, 2014))
years
```

```python
df_can[(df_can['Continent']=='Asia') &(df_can['Region']=='Southern Asia')]
```

```python
print('Data dimensions:', df_can.shape)
print(df_can.columns)
df_can.head(3)
```

```python
%matplotlib inline 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
haiti = df_can.loc['Haiti', years]
haiti
```

```python
haiti.plot()
```

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind='area')

plt.title('Immigration from Haiti to Canada')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.text(2000, 6000, '2010 Earthquake') 
plt.savefig('haiti_immigration.png')
plt.show()
```

```python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = df_can.head(5)
df_top5 = df_top5[years].transpose()
print(df_top5)
```

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='area', figsize=(15, 8))

plt.title('Immigration Trend of Top 5 Countries to Canada')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.savefig('top5_immigration.png')
plt.show()
```
